## Objective -

The Objectives of this monthly spendings dataset is to analyze how monthly expenses vary across different categories such as groceries, rent, transportation, healthcare, utilities, and entertainment.

To identify which categories contribute the most to total monthly spending.

To visualize monthly spending trends and detect increases or decreases in expenses over time.

To measure how savings and investments change month by month.

To create visualizations (bar charts, line plots, Histograph) for a clear representation of financial data.

To communicate insights effectively through dashboards or reports.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
Ct.columns = Ct.columns.str.strip()      # removes spaces
Ct.columns = Ct.columns.str.lower()      # makes all lowercase
print(Ct.columns.tolist())               # check again
```

```python
Ct['month'] = pd.to_datetime(Ct['month'])   # convert to datetime if not already

Ct_2020 = Ct[Ct['month'].dt.year == 2020]
Ct_2021 = Ct[Ct['month'].dt.year == 2021]
Ct_2022 = Ct[Ct['month'].dt.year == 2022]
Ct_2023 = Ct[Ct['month'].dt.year == 2023]
Ct_2024 = Ct[Ct['month'].dt.year == 2024]
Ct_2025 = Ct[Ct['month'].dt.year == 2025]
```

```python
total_groceries_2020 = Ct_2020['groceries (₹)'].sum()
print("Total groceries in 2020:", total_groceries_2020)

total_groceries_2021 = Ct_2021['groceries (₹)'].sum()
print("Total groceries in 2021:", total_groceries_2021)

total_groceries_2022 = Ct_2022['groceries (₹)'].sum()
print("Total groceries in 2022:", total_groceries_2022)

total_groceries_2023 = Ct_2023['groceries (₹)'].sum()
print("Total groceries in 2023:", total_groceries_2023)

total_groceries_2024 = Ct_2024['groceries (₹)'].sum()
print("Total groceries in 2024:", total_groceries_2024)

total_groceries_2025 = Ct_2025['groceries (₹)'].sum()
print("Total groceries in 2025:", total_groceries_2025)

```

## Data Analysis And Data Visualization

```python
plt.figure(figsize=(10,5))
plt.plot(Ct['month'], Ct['savings (₹)'], marker='o', color='c')
plt.title("Monthly Savings Trend")
plt.xlabel('Month')
plt.ylabel('Savings (₹)')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
Monthly_healthcare = Ct.groupby('month')['healthcare (₹)'].sum().reset_index()

plt.Figure(figsize=(25,20))
sns.barplot(data=Monthly_healthcare, x = "month" , y = "healthcare (₹)" ,palette='viridis')
plt.title("Total healthcare spending by month")
plt.xticks(rotation = 90)
plt.yticks(rotation = 90)
plt.show()
```

```python
Monthly_healthcare_2025 = Ct_2025.groupby(Ct_2025['month'].dt.month)['healthcare (₹)'].sum().reset_index()

plt.Figure(figsize=(25,20))
sns.barplot(data=Monthly_healthcare_2025, x = "month" , y = "healthcare (₹)" ,palette='viridis')
plt.title("Total healthcare spending by month 2025")
plt.xticks(rotation = 90)
plt.yticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(Ct["income (₹)"], bins = 10, kde = True, color = 'purple')
plt.title("income histogram")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=Ct, x='month', y='income (₹)', hue='savings (₹)', style='emi/loans (₹)', s=100)
plt.title('monthly income')
plt.show()
```
